```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion, size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()
    
def spectogram(data, sr, emotion):
     x = librosa.stft(data)
     xdb = librosa.amplitude_to_db(abs(x))
     plt.figure(figsize=(11,4))
     plt.title(emotion, size=20)
     librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
     plt.colorbar()
```

### Feature Extraction

```python
def extract_mfcc(filename):
     data, sr = librosa.load(filename, duration=3, offset=0.5)
     mfcc = np.mean(librosa.feature.mfcc(y=data, sr=sr, n_mfcc=40).T, axis=0)
     return mfcc
```

```python
extract_mfcc(df['FilePaths'][0])
```

```python
X_mfcc = df['FilePaths'].apply(lambda x: extract_mfcc(x))
```

```python
X_mfcc
```

```python
#One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
y=encoder.fit_transform(df[['Labels']])
y
```

```python
y=y.toarray()
y
```

```python
y.shape
```

### Creating LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
model=Sequential([
    LSTM(256, return_sequences=True, input_shape=(40,1)),
    Dropout(0.2),
    LSTM(128, return_sequences=False),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
clf=model.fit(X,y, validation_split=0.2, epochs=100, batch_size=512, shuffle=True)
```
